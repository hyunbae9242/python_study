```python
import urllib.request as request
import urllib.parse as parse

# https://www.google.com/search?q=python&ie=UTF-8
api = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
value = {"stnId": 108}
params = parse.urlencode(value)
url = api + "?" + params
data = request.urlopen(url).read()
text = data.decode("utf-8")
print(text)

```

```python
import urllib.request as request
import urllib.parse as parse

regionNumber = int(input("지역번호를 입력하세요."))
api = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
value = {"stnId": regionNumber}
params = parse.urlencode(value)
url = api + "?" + params
data = request.urlopen(url).read()
text = data.decode("utf-8")
print(text)

```

```python
from bs4 import BeautifulSoup

text = """
<html><body>
<div id="meigen">
    <h1>위키북스 도서 </h1>
    <ul class="items">
        <li>유니티 게임 이펙트 입문</li>
        <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
        <li>모던 웹사이트 디자인의 정석</li>
    </ul>
</div>
</body></html>
"""
soup = BeautifulSoup(text, 'html.parser')

h1 = soup.html.body.div.h1
print(h1)
h2 = soup.find("h1")
print(h2)
print(h2.string)
h3 = soup.find("h1").string
print(h3)
li2 = soup.select_one("div > ul > li").string
print(li2)
```
